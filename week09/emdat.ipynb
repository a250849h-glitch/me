# データの読み込み

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
emdat = pd.read_csv('emdat.csv')
```

```python
# Explore the data
emdat.head()  # Display the first few rows
```

# 分析

```python
# Ranking by total deaths
death_ranking = emdat.groupby('Country')['Total Deaths'].sum().sort_values(ascending=False)
print("Top 10 countries by total deaths from disasters:")
print(death_ranking.head(10))
```

```python
# ranking of total deaths in horizontal bar chart, top 10 countries, highest first 
top_10_deaths = death_ranking.head(10)
plt.figure(figsize=(10, 6))
top_10_deaths.plot(kind='barh', color='red')
plt.xlabel('Total Deaths')
plt.title('Top 10 Countries by Total Deaths from Disasters')
plt.gca().invert_yaxis()  # Highest first
plt.show()
```

```python
# disasters by year
disasters_by_year = emdat.groupby('Start Year').size()
plt.figure(figsize=(10, 6))
disasters_by_year.plot(kind='bar', color='blue')

# label only every 10 years on x-axis
plt.xticks(ticks=range(0, len(disasters_by_year), 10), labels=disasters_by_year.index[::10])   
plt.xlabel('Start Year')
plt.ylabel('Number of Disasters')
plt.title('Number of Disasters by Start Year')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV読み込み
df = pd.read_csv('emdat.csv')

# ==============================
# 1. 災害タイプ別 件数
# ==============================
counts = df['Disaster Type'].value_counts()

plt.figure()
counts.plot(kind='bar')
plt.title('Disaster Type - Counts')
plt.xlabel('Disaster Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# ==============================
# 2. 災害タイプ別 死亡者数合計
# ==============================
deaths = df.groupby('Disaster Type')['Total Deaths'].sum().sort_values(ascending=False)

plt.figure()
deaths.plot(kind='bar')
plt.title('Disaster Type - Total Deaths')
plt.xlabel('Disaster Type')
plt.ylabel('Total Deaths')
plt.tight_layout()
plt.show()

# ==============================
# 3. 国ごとの災害発生件数 Top10
# ==============================
country_counts = df['Country'].value_counts().head(10)

plt.figure()
country_counts.plot(kind='bar')
plt.title('Top 10 Countries by Disaster Count')
plt.xlabel('Country')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ====================================
# Load CSV
# ====================================
df = pd.read_csv("emdat.csv")

# ====================================
# 1. Bar chart: Count of Disaster Types
# ====================================
counts = df["Disaster Type"].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(counts.index, counts.values)
plt.xticks(rotation=45, ha='right')
plt.title("Number of Disasters by Type")
plt.xlabel("Disaster Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ====================================
# 2. Heatmap: Disaster Type × Decade
# ====================================
df["Decade"] = (df["Start Year"] // 10) * 10
pivot = df.pivot_table(
    index="Disaster Type",
    columns="Decade",
    aggfunc="size",
    fill_value=0
)

plt.figure(figsize=(10, 6))
plt.imshow(pivot, aspect='auto')
plt.colorbar(label="Count")
plt.title("Disaster Type × Decade Heatmap")
plt.xlabel("Decade")
plt.ylabel("Disaster Type")
plt.xticks(
    ticks=np.arange(len(pivot.columns)),
    labels=pivot.columns,
    rotation=45
)
plt.yticks(
    ticks=np.arange(len(pivot.index)),
    labels=pivot.index
)
plt.tight_layout()
plt.show()

# ====================================
# 3. Plot disasters on a world map (lat/long)
# ====================================
df_map = df.dropna(subset=["Latitude", "Longitude"])  # remove rows without coordinates

# ---- Install cartopy if necessary ----
# !pip install cartopy

import cartopy.crs as ccrs
import cartopy.feature as cfeature

plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_global()
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.add_feature(cfeature.LAND, alpha=0.3)

# scatter plot
ax.scatter(
    df_map["Longitude"],
    df_map["Latitude"],
    s=20,
    color="red",
    alpha=0.7,
    transform=ccrs.PlateCarree(),
    label="Disaster"
)

plt.title("Disaster Locations (Latitude / Longitude)")
plt.legend()
plt.show()
```

```python
import pandas as pd
import folium

df = pd.read_csv("emdat.csv")
df_map = df.dropna(subset=["Latitude", "Longitude"])

center_lat = df_map["Latitude"].mean()
center_lon = df_map["Longitude"].mean()

m = folium.Map(location=[center_lat, center_lon], zoom_start=2)

for _, row in df_map.iterrows():
    folium.CircleMarker(
        location=[row["Latitude"], row["Longitude"]],
        radius=4,
        color="red",
        fill=True,
        fill_color="red",
        tooltip=f"{row['Disaster Type']} ({row['Start Year']})"
    ).add_to(m)

# VSCodeではこれが絶対必要
m.save("disaster_map.html")
print("地図を 'disaster_map.html' として保存しました。VSCode の左側に表示されます。")
```

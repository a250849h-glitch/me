# 幸福度の差がある要素を比較

# まずは高い国の要素を調べる

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm

# フォント設定（macOS対応）
plt.rcParams["font.family"] = "Hiragino Sans"
plt.rcParams["axes.unicode_minus"] = False

# データ読み込み
df = pd.read_csv("world_happiness_report.csv")

# 2015年・幸福度上位5か国
top = (
    df[df["year"] == 2015]
    .sort_values("Happiness Score", ascending=False)
    .head(5)
)

factors = [
    "Economy (GDP per Capita)",
    "Family",
    "Health (Life Expectancy)",
    "Freedom",
    "Trust (Government Corruption)",
    "Generosity"
]

values = top[factors].mean()

# レーダーチャート準備
labels = factors
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False)
values = np.concatenate([values, [values[0]]])
angles = np.concatenate([angles, [angles[0]]])

# 描画
plt.figure(figsize=(7, 7))
ax = plt.subplot(111, polar=True)
ax.plot(angles, values)
ax.fill(angles, values, alpha=0.25)

ax.set_thetagrids(angles[:-1] * 180 / np.pi, labels)
plt.title("幸福度が高い国の特徴（2015年・上位5か国平均）")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 日本語フォント（macOS）
plt.rcParams["font.family"] = "Hiragino Sans"
plt.rcParams["axes.unicode_minus"] = False

df = pd.read_csv("world_happiness_report.csv")

# 2015年のみ
df_2015 = df[df["year"] == 2015]

# 幸福度 上位5・下位5
top = df_2015.sort_values("Happiness Score", ascending=False).head(5)
bottom = df_2015.sort_values("Happiness Score", ascending=True).head(5)

factors = [
    "Economy (GDP per Capita)",
    "Family",
    "Health (Life Expectancy)",
    "Freedom",
    "Trust (Government Corruption)",
    "Generosity"
]

top_mean = top[factors].mean()
bottom_mean = bottom[factors].mean()
```

# 棒グラフにしてわかりやすく比較します


```python
x = np.arange(len(factors))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, top_mean, width, label="幸福度が高い国")
plt.bar(x + width/2, bottom_mean, width, label="幸福度が低い国")

plt.xticks(x, factors, rotation=45, ha="right")
plt.ylabel("平均値")
plt.title("幸福度が高い国と低い国の要因比較（2015年）")
plt.legend()
plt.tight_layout()
plt.show()
```

# レーダーチャートでも比較してみる

```python
labels = factors
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False)

top_values = np.concatenate([top_mean, [top_mean[0]]])
bottom_values = np.concatenate([bottom_mean, [bottom_mean[0]]])
angles = np.concatenate([angles, [angles[0]]])

plt.figure(figsize=(7, 7))
ax = plt.subplot(111, polar=True)

ax.plot(angles, top_values, label="幸福度が高い国")
ax.fill(angles, top_values, alpha=0.25)

ax.plot(angles, bottom_values, label="幸福度が低い国")
ax.fill(angles, bottom_values, alpha=0.25)

ax.set_thetagrids(angles[:-1] * 180 / np.pi, labels)
plt.title("幸福度が高い国と低い国の特徴比較（2015年）")
plt.legend(loc="upper right")
plt.show()
```

このグラフから家族、経済、健康の面での要因が高いことがわかる。また、寛大さはそこまで差はない。

# 発見したこと（Results）
本分析から、以下の点が明らかになった。

幸福度が高い国は、経済力（GDP）、健康（平均寿命）、社会的支援（Family）の水準が一様に高い

→ 特定の要因だけでなく、複数の指標が同時に高いことが確認された。

幸福度が低い国では、特に経済力と健康指標が著しく低い

→ レーダーチャートにおいても、これらの項目が大きく内側に位置していた。

自由度や政府への信頼も、幸福度の高低を分ける重要な要因である

→ 上位国では自由度・汚職の少なさが比較的高く、制度面の安定が示唆された。

Generosity（寛容さ）は他の要因と比べて差が小さい

→ 幸福度を左右する主因ではあるが、経済・健康ほどの影響力は見られなかった。

# 考察（Discussion）
幸福度が高い国では、国民が安定した収入を得られる経済基盤と、長期的に健康を維持できる医療・生活環境が整っている。このことは、生活の不安を軽減し、将来への見通しを持ちやすくするため、幸福度の向上につながると考えられる。
また、社会的支援（Family）が高いことは、人間関係やコミュニティの安定を示しており、精神的な安心感を与える要因である。さらに、自由度や政府への信頼が高い国では、制度への不満や不安が少なく、個人が自律的に行動しやすい環境が整っていると推察される。
一方、幸福度が低い国では、経済的困難や医療体制の不十分さに加え、制度への不信が重なり、生活全体の満足度を下げている可能性が高い。これらの結果から、幸福度向上には経済成長のみならず、健康、社会的支援、制度の質を含めた包括的な政策が必要であると考えられる。

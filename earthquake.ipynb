```python
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.geojson"
data = requests.get(url).json()
```

# === üìä „Éá„Éº„ÇøÊï¥ÂΩ¢ ===
# featuresÈÉ®ÂàÜ„ÇíÂèñ„ÇäÂá∫„Åó„Å¶„ÄÅÂú∞Èúá„ÅÆÂü∫Êú¨ÊÉÖÂ†±„ÇíDataFrameÂåñ

```python
features = data["features"]
earthquakes = pd.DataFrame([{
    "place": f["properties"]["place"],
    "magnitude": f["properties"]["mag"],
    "time": pd.to_datetime(f["properties"]["time"], unit="ms"),
    "longitude": f["geometry"]["coordinates"][0],
    "latitude": f["geometry"]["coordinates"][1],
    "depth": f["geometry"]["coordinates"][2]
} for f in features])
```

# === üó∫Ô∏è Folium„ÅßÂú∞Âõ≥„Çí‰ΩúÊàê ===

```python
m = folium.Map(location=[0, 0], zoom_start=2)
```

# ÂêÑÂú∞Èúá„ÇíÂú∞Âõ≥„Å´ËøΩÂä†

```python
for _, row in earthquakes.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=max(row["magnitude"], 1) * 2,  # Âú∞Èúá„ÅÆË¶èÊ®°„Åß„Éû„Éº„Ç´„Éº„ÅÆÂ§ß„Åç„Åï„ÇíÂ§â„Åà„Çã
        color="red" if row["magnitude"] >= 5 else "orange",
        fill=True,
        fill_opacity=0.7,
        popup=f"{row['place']} (M{row['magnitude']})"
    ).add_to(m)
```

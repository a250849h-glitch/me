```
import requests, pandas as pd, plotly.express as px
url = 'https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_week.geojson'
data = requests.get(url).json()
features = data['features']
df = pd.json_normalize(features)
df = df[['properties.mag', 'properties.place', 'geometry.coordinates']]
df.columns = ['magnitude', 'place', 'coordinates']
fig = px.scatter_geo(df,
    lon=df['coordinates'].apply(lambda x: x[0]),
    lat=df['coordinates'].apply(lambda x: x[1]),
    color='magnitude', size='magnitude',
    title='Global Earthquake Impact (USGS, Past Week)',
    color_continuous_scale='turbo'
)
fig.show()
```
